### Задание №2
Написать генератор случайных матриц(многомерных), который принимает
опциональные параметры <b>m</b>, <b>n</b>, <b>min_limit</b>, <b>max_limit</b>, где <b>m</b> и <b>n</b> указывают размер
матрицы, а <b>min_lim</b> и <b>max_lim</b> - минимальное и максимальное значение для
генерируемого числа.

```python
import random
import time
```

```python
import random
 
user_m = int(input('Введите количество строк: '))
user_n = int(input('Введите количество столбцов: '))
user_min_limit = int(input('Введите минимальный порог значений: '))
user_max_limit = int(input('Введите максимальный порог значений: '))
def matrixGen(m, n, min_limit, max_limit):
  matrix = []
  for i in range(m):
    arr = []
    for j in range(n):
      arr.append(random.randint(min_limit, max_limit))
    matrix.append(arr)
  return matrix
matr = matrixGen(user_m, user_n, user_min_limit, user_max_limit)
for row in matr:
  print(row)
```

### Задание №3
Реализовать методы сортировки строк числовой матрицы в соответствии с
заданием. Оценить время работы каждого алгоритма сортировки и сравнить его со
временем стандартной функции сортировки. Испытания проводить на сгенерированных
матрицах.

```python
import copy
```

```python
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.

import copy
import time

def SelectionSort(matrix):
  for array in matrix:
    for i in range(len(array) - 1):
      m = i
      j = i + 1
      while j < len(array):
          if array[j] < array[m]:
              m = j
          j = j + 1
      array[i], array[m] = array[m], array[i]
  return matrix

matr1 = deepcopy(matr)
start_time = time.time()
SelectionSort(matr1)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
for row in matr1:
  print(row)
```

```python
# Сортировка вставкой.

import time
import copy

def InsertionSort(matrix):
  for array in matrix:
    for i in range(1, len(array)):
      key = array[i]
      j = i-1
      while j >=0 and key < array[j] :
          array[j+1] = array[j]
          j -= 1
      array[j+1] = key 

matr1 = copy.deepcopy(matr)
start_time = time.time()
InsertionSort(matr1)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
for row in matr1:
  print(row)
```

```python
# Сортировка обменом. || Сортировка пузырьком.

import copy
import time

def BubbleSort(matrix):
  for array in matrix:
    for i in range(len(array)-1):
        for j in range(len(array)-i-1):
            if array[j] > array[j+1]:
                buff = array[j]
                array[j] = array[j+1]
                array[j+1] = buff

matr1 = deepcopy(matr)
start_time = time.time()
BubbleSort(matr1)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
for row in matr1:
  print(row)
```

```python
from IPython.core.display import deepcopy
# Сортировка Шелла.

import time
import copy
import math

def ShellSort(matrix):
  for array in matrix:
     n = len(array)
     k = int(math.log2(n))
     interval = 2**k -1
     while interval > 0:
        for i in range(interval, n):
            temp = array[i]
            j = i
            while j >= interval and array[j - interval] > temp:
                array[j] = array[j - interval]
                j -= interval
            array[j] = temp
        k -= 1
        interval = 2**k -1

matr1 = deepcopy(matr)
start_time = time.time()
ShellSort(matr1)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
for row in matr1:
  print(row)
```

```python
# Быстрая сортировка.

import time 
import copy

matr1 = deepcopy(matr)
start_time = time.time()
for array in matr1:
  array.sort()
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
for row in matr1:
  print(row)
```

```python
# Турнирная сортировка (Пирамидальная сортировка)

import time
import copy

# Для реализации пирамидальной сортировки потребуется вспомогательная функция:
def heapify(arr, n, i):
    largest = i 
    l = 2 * i + 1
    r = 2 * i + 2 
    if l < n and arr[i] < arr[l]:
        largest = l
    if r < n and arr[largest] < arr[r]:
        largest = r

    if largest != i:
        arr[i],arr[largest] = arr[largest],arr[i]

        heapify(arr, n, largest)

def heapSort(arr):
    n = len(arr)

    for i in range(n, -1, -1):
        heapify(arr, n, i)

    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i] # свап 
        heapify(arr, i, 0)

matr1 = deepcopy(matr)
start_time = time.time()
for array in matr1:
  heapSort(array)

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
for row in matr1:
  print(row)
```
